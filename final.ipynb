```python
import pandas as pd

df = pd.read_csv('cve.csv')
```

```python
df.head()
```

```python
df.head()
```

```python
df['severity_level'] = pd.cut(df['cvss'], bins=[0,4, 7, float('inf')], labels=['low','medium', 'high'])
label = df['severity_level']
```

```python
df.head()
```

```python
columns = ['cv_number','severity_level','cwe_code']
```

columns = []

```python
df = df[columns]
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Label encode cv_number and severity
le_cv = LabelEncoder()
df['cv_number'] = le_cv.fit_transform(df['cv_number'])

le_severity = LabelEncoder()
df['severity_level'] = le_severity.fit_transform(df['severity_level'])
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

```

```python
X = df[['cv_number', 'cwe_code']]  # Assuming these are features
y = df['severity_level']  # This is the label to predict
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

y_pred = model.predict(X_test)


# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')  # Weighted average for imbalanced datasets
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

# Print the metrics
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-Score:", f1)
```

# RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier

# Build a random forest model
model1 = RandomForestClassifier(n_estimators=100)  # Adjust hyperparameters as needed

model1.fit(X_train, y_train)

y_pred1 = model1.predict(X_test)

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred1)
precision = precision_score(y_test, y_pred1, average='weighted')  # Weighted average for imbalanced datasets
recall = recall_score(y_test, y_pred1, average='weighted')
f1 = f1_score(y_test, y_pred1, average='weighted')

# Print the metrics
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-Score:", f1)
```

# SVC

```python
from sklearn.svm import SVC

# Build an SVC model
model2 = SVC()  # Adjust hyperparameters as needed

model2.fit(X_train, y_train)

y_pred2 = model2.predict(X_test)

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred2)
precision = precision_score(y_test, y_pred2, average='weighted')  # Weighted average for imbalanced datasets
recall = recall_score(y_test, y_pred2, average='weighted')
f1 = f1_score(y_test, y_pred2, average='weighted')

# Print the metrics
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-Score:", f1)
```
